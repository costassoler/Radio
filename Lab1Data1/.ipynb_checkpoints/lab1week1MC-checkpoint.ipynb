```python
def plot_waveform(time, filename):
    
    """Plots and saves the plot as a pdf for our Nyquist sampling
    data.
            time: array of time values to use on the x-axis in microseconds
            filename: name of the txt file where the voltage data is stored. """
    
    trialnum = int(filename[-1])
    data = np.loadtxt(filename)
    plt.plot((time[:30])*10e6, np.loadtxt(filename)[:30], 'k.-', linewidth = 2.0, markersize = 10)
    plt.title('Trial ' + str(trialnum) + ' : ' + str(31.25*trialnum/10) + ' MHz')
    plt.ylabel('Voltage (mV)')
    plt.xlabel('Time (microsec)')
    plt.rc('font', size=18)
    plt.rc('axes', titlesize=18)
    plt.rc('font',**{'family':'serif','serif':['Times']})
    plt.rc('text', usetex=True)
    plt.savefig(filename + 'plot.pdf', bbox_inches = "tight")
    plt.show()
```

## Voltage Spectra

```python
def voltage_spec(filename, x):
    v = ugradio.dft.dft(np.loadtxt(filename), vsamp = 31.25e6)
    real = v[1].real
    imaginary = v[1].imag
    plt.plot(v[0], real, 'b-.', label='real')
    plt.plot(v[0], imaginary, 'g-', label='imaginary')
    plt.title('Voltage Spectrum Trial: ' + str(x))
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Voltage')
    plt.legend()
    plt.show()
```

## Nyquist Windows

```python
def window(array, n):
    big_freq = np.linspace(n*(31.25e6)/2, -1*n*(31.25e6)/2, num=10000)
    big_fourier = ugradio.dft.dft(array, f = big_freq, vsamp = 31.25e6)
    real = big_fourier[1].real
    imaginary = big_fourier[1].imag
    plt.plot(big_fourier[0], real, 'b-.', label='real')
    plt.plot(big_fourier[0], imaginary, 'g-', label='imaginary')
    plt.title('Voltage Spectrum of Nyquist Window, N = ' + str(n))
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Voltage')
    plt.legend()
    plt.show()
    big_P = np.abs(big_fourier[1])**2
    plt.plot(big_fourier[0], big_P)
    plt.title('Power Spectrum of Nyquist Window, N = '+str(n))
    plt.ylabel('Power')
    plt.xlabel('Frequency (Hz)')
    plt.show()
```

```python
def noise_capture(number):
    noise_data = ugradio.pico.capture_data('1V')
    noise_filename = 'Noise'+str(number)
    np.savetxt(noise_filename, noise_data)
```

```python
def power_spec(filename):
    dft = ugradio.dft.dft(filename, vsamp = 31.25e6)
    first = dft[0]
    power = abs(dft[1])**2
    return first, power
```

```python
from scipy.optimize import curve_fit
```

```python
#SPREAD
def spread(x):
    n = len(two_blocks[1])
    xbar = np.mean(thirty_two_blocks[1])
    std = np.sqrt(np.sum((x - xbar)**2)/n)
    return xbar/std
```

```python
spread_2 = spread(two_blocks[1])
spread_4 = spread(four_blocks[1])
spread_8 = spread(eight_blocks[1])
spread_16 = spread(sixteen_blocks[1])
```

```python
def n(x, a, b):
    x = x.astype(float)
    return a*(x**(b))
x = np.array([2, 4, 8, 16])
```

```python
opt, cov = curve_fit(n, x, np.array([spread_2, spread_4, spread_8, spread_16]))
```

```python
plt.plot(np.array([2, 4, 8, 16]), np.array([spread_2, spread_4, spread_8, spread_16]), 'bo', label = 'SNR')
plt.plot(x, n(x, opt[0], opt[1]), 'r-', label = '\r$N^{0.11}$')
plt.legend(loc = 'lower right', prop = {'size' : 12})
plt.title('SNR vs Number of Blocks Averaged')
plt.xlabel('Number of Blocks Averaged (N)')
plt.ylabel('Signal to Noise Ratio (SNR)')
plt.xlim(1, 17)
plt.savefig('SNR.pdf', bbox_inches = "tight")
plt.show()
```

```python
def acf(data):
    total_acf = np.array([])
    for dn in range(0, 2000):
        num = 0
        den = 0
        for i in range(0, 16000 - dn - 1):
            num += data[i]*data[i + dn]
            den += data[i]**2
        total_acf = np.append(total_acf, num/den)
    return total_acf
```

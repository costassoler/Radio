## Voltage Spectra

```python
def voltage_spec(filename, x):
    v = ugradio.dft.dft(np.loadtxt(filename), vsamp = 31.25e6)
    real = v[1].real
    imaginary = v[1].imag
    plt.plot(v[0], real, 'b-.', label='real')
    plt.plot(v[0], imaginary, 'g-', label='imaginary')
    plt.title('Voltage Spectrum Trial: ' + str(x))
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Voltage')
    plt.legend()
    plt.show()
```

## Nyquist Windows

```python
def window(array, n):
    big_freq = np.linspace(n*(31.25e6)/2, -1*n*(31.25e6)/2, num=10000)
    big_fourier = ugradio.dft.dft(array, f = big_freq, vsamp = 31.25e6)
    real = big_fourier[1].real
    imaginary = big_fourier[1].imag
    plt.plot(big_fourier[0], real, 'b-.', label='real')
    plt.plot(big_fourier[0], imaginary, 'g-', label='imaginary')
    plt.title('Voltage Spectrum of Nyquist Window, N = ' + str(n))
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Voltage')
    plt.legend()
    plt.show()
    big_P = np.abs(big_fourier[1])**2
    plt.plot(big_fourier[0], big_P)
    plt.title('Power Spectrum of Nyquist Window, N = '+str(n))
    plt.ylabel('Power')
    plt.xlabel('Frequency (Hz)')
    plt.show()
```

```python
def acf(data):
    total_acf = np.array([])
    for dn in range(0, 2000):
        num = 0
        den = 0
        for i in range(0, 16000 - dn - 1):
            num += data[i]*data[i + dn]
            den += data[i]**2
        total_acf = np.append(total_acf, num/den)
    return total_acf
```

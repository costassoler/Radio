## Voltage Spectra

```python
def voltage_spec(filename, x):
    v = ugradio.dft.dft(np.loadtxt(filename), vsamp = 31.25e6)
    real = v[1].real
    imaginary = v[1].imag
    plt.plot(v[0], real, 'b-.', label='real')
    plt.plot(v[0], imaginary, 'g-', label='imaginary')
    plt.title('Voltage Spectrum Trial: ' + str(x))
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Voltage')
    plt.legend()
    plt.show()
```

```python
voltage_spec('trial0.1', 1)
```

```python
voltage_spec('trial0.2', 2)
```

```python
voltage_spec('trial0.3', 3)
```

```python
voltage_spec('trial0.4', 4)
```

```python
voltage_spec('trial0.5', 5)
```

```python
voltage_spec('trial0.6', 6)
```

```python
voltage_spec('trial0.7', 7)
```
